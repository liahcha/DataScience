## Installation Guide

1. CUDA / cuDNN installation & path configuration
2. Install Tensorflow (GPU version)
3. Install Keras (with Tensorflow backend)


## CUDA / cuDNN Installation

### 1. Check if CUDA is installed correctly
  > cd /usr/localcuda-8.0/samples/1_Utilities/bandwidthTest/ <br/>
  > sudo make <br/>
  > ./bandwidthTest <br/>
  
  ![](./img/01_cuda_01.png)
  
<br/>

  > cd /usr/local/cuda-8.0/sampels/1_Utilities/deviceQuery/ <br/>
  > sudo make <br/>
  > ./deviceQuery <br/>
  
  ![](./img/01_cuda_02.png)
  
<br/>
  * Both tests above must show "RESULT=PASS" 


### 2. Configure 'PATH' and 'LD_LIBRARY_PATY'

   * 환경변수 확인
   > echo `$` PATH <br/>
   > echo `$` LD_LIBRARY_PATH <br/>
   
   ![](./img/01_cuda_03.png)
   
   <br/>
   
   * gedit을 사용하여 .bashrc 파일 열기
   > gedit ~/.bashrc 
   
   <br/>
   
   * bashrc 파일 내 다음과 같이 추가 후 저장
   
   > export PATH="/usr/local/cuda-8.0/bin:`$`PATH" <br/>
   > export LD_LIBRARY_PATH="/usr/local/cuda-8.0/lib64:`$`LD_LIBRARY_PATH" <br/>
   
      ![](./img/01_cuda_05.png)
      
   <br/>
   
   > Reload './bashrc'file
   <br/>
   
   * 환경변수 재확인
   
     ![](./img/01_cuda_04.png)
     
   <br/><br/>  

### 3. Install & Configure cuDNN v5.1
  * Available at https://developer.nvidia.com/rdp/cudnn-download
  
     ![](./img/01_cuda_06.png)
     
  <br/>
  * cuDNN 파일이 존재하는 디렉토리로 이동
  > cd ~/다운로드
  
  * ~/다운로드/cuda 파일이 생성됨 (압축풀기)
  > tar xzvf cudnn-8.0-linux-x64-v5.1.tgz
  
  * cuDNN 파일을 cuda-8.0 디렉토리로 이동
  > sudo cp cuda/lib64/* /usr/local/cuda-8.0/lib64 <br/>
  > sudo cp cuda/include/* /usr/local/cuda-8.0/include/ <br/>
  
    ![](./img/01_cuda_07.png)
    
    <br/><br/>
    
## Tensorflow installation (with Tensorflow backend)

In the terminal,

1. Python 버전확인 (/home/user/anaconda3/bin/python)
> which python

2. Tensorflow GPU 버전 설치
> pip install tensorflow-gpu
![](./img/01_cuda_08.png)

3. Keras 설치 (이미 설치되어 있을 경우 불필요)
> pip install keras
![](./img/01_cuda_09.png)
